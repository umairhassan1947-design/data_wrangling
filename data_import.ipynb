Reading dummy data in python
- find top 5 countires with most custometrs
- top 5 wesbites conversion and its frequency
- visualise using barchart
- How many people have last name 'Callahan'


```python
import pandas as pd
df = pd.read_csv('customers.csv')
topcountries = df["Country"].value_counts().head(5)
print(topcountries)
topwebsites = df["Website"].value_counts().head(5)
print(topwebsites)

import matplotlib.pyplot as plt
topcountries.plot(kind= "bar",title = "Top 5 countries and their frequency")
plt.ylabel("Number of customers")
plt.show()
topwebsites.plot(kind="bar",title = "top 5 websites converted")
plt.ylabel("Number of customers")
plt.show()

callahan_count = (df["Last Name"] == "Callahan").sum()
print('Number of customers with the name Callahan is ', callahan_count)

df
```
